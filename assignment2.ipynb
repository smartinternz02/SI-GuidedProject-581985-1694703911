#MULTI VARIATE

      Inference:

#RELATION PLOT

```python
sns.relplot(x="alcohol", y="speeding", hue="ins_premium", data=df)
plt.title("Relationship between Alcohol and Speeding with Insurance Premium")
plt.xlabel("Alcohol Involvement (%)")
plt.ylabel("Speeding Incidents")
plt.show()
```

      #INFERENCE:
      The relational plot illustrates a potential association between higher alcohol involvement and increased speeding incidents.
      Variations in insurance premium costs suggest that regions with elevated alcohol involvement may face diverse insurance pricing.

```python
# Create a box plot of 'total' accidents grouped by distracted driving status
plt.figure(figsize=(8, 6))
sns.boxplot(x="not_distracted", y="total", data=df)
plt.xlabel("Distracted Driving")
plt.ylabel("Total Accidents")
plt.title("Box Plot of Total Accidents by Distracted Driving Status")
plt.show()
```

      # Inference:
      The boxplot between "not_distracted" and "total" accidents in the car_crashes dataset reveals that regions with higher levels
      of not being distracted tend to have a wider range of total accidents.

#JOINT PLOT

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x="abbrev", y="ins_premium")
plt.xlabel("Region (Abbreviation)")
plt.ylabel("Insurance Premium")
plt.title("Box Plot of Insurance Premium by Region")
plt.xticks(rotation=45)
plt.show()
```

      The joint plot between "Total Accidents" and "Insurance Losses" suggests a potential positive relationship.
      As the total number of accidents (x-axis) increases, there seems to be a tendency for higher insurance losses (y-axis).
      However, the relationship is not strictly linear, indicating that while there's a general trend of higher accidents leading
      to higher losses, it's not perfectly proportional and may be influenced by other factors.

#HISTOGRAM

```python
# Create a histogram for the "Total Accidents" column
plt.hist(df['total'], bins=10, edgecolor='k')
plt.title('Histogram of Total Accidents')
plt.xlabel('Total Accidents')
plt.ylabel('Frequency')
plt.grid(True)

plt.show()
```

      The histogram of "Total Accidents" shows that the majority of states have a relatively low number of accidents, with a peak around 15-17 accidents.
      A few states exhibit higher accident rates as indicated by the right tail of the distribution.
      This distribution provides insights into the variation in accident frequencies among states
